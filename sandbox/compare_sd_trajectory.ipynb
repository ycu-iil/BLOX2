```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

cutoff = 0

plt.rcParams.update({
    "font.size": 16,
    "axes.labelsize": 18,
    "axes.titlesize": 18,
    "xtick.labelsize": 16,
    "ytick.labelsize": 16,
    "legend.fontsize": 16,
})

csv_paths = [
    "results/sd_rf_nes10_s1.csv",
    "results/sd_rf_nes10_s1_batch10.csv",
]
names = [
    "Batch size: 1 (disabled)",
    "Batch size: 10",
]

plt.figure()

for path, name in zip(csv_paths, names):
    y = pd.read_csv(path, header=None).iloc[:, 0].to_numpy()[cutoff:]
    x = range(cutoff + 1, len(y) + cutoff + 1)
    plt.plot(x, y, label=name)
    
plt.xscale("log", base=10)
plt.grid()
plt.xlabel("Number of samplings")
plt.ylabel("Stein discrepancy")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Compare selection time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

cutoff = 0

plt.rcParams.update({
    "font.size": 16,
    "axes.labelsize": 18,
    "axes.titlesize": 18,
    "xtick.labelsize": 16,
    "ytick.labelsize": 16,
    "legend.fontsize": 16,
})

csv_paths = [
    "results/time_orig_rf_nes100_s1.csv",
    "results/time_rf_nes100_s1_level02.csv",
]
names = [
    "BLOX (original)",
    "BLOX2",
]

plt.figure()

for path, name in zip(csv_paths, names):
    y = pd.read_csv(path)["Selection"]
    x = range(cutoff + 1, len(y) + cutoff + 1)
    plt.plot(x, y, label=name)
    
# plt.yscale("log", base=10)
plt.grid()
plt.xlabel("Number of samplings")
plt.ylabel("Selection time")
plt.legend()
plt.tight_layout()
plt.show()
```

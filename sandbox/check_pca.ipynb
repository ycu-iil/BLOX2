```python
import pandas as pd
from blox2 import load_features

X = load_features("../data/zinc_dft/feature_list_mfp2048_rdd.npz")
# X = load_features("../data/material/features.csv")
X
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

pca = PCA(svd_solver="randomized", random_state=0)
X_pca = pca.fit_transform(X_std)

explained_ratio = pca.explained_variance_ratio_
cumulative_ratio = np.cumsum(explained_ratio)
```

```python
plt.rcParams.update({
    "font.size": 16,
    "axes.labelsize": 18,
    "axes.titlesize": 18,
    "xtick.labelsize": 16,
    "ytick.labelsize": 16,
    "legend.fontsize": 16,
})

plt.figure(figsize=(7, 5))
plt.plot(
    np.arange(1, len(cumulative_ratio) + 1),
    cumulative_ratio,
)
plt.xlabel("Number of components")
plt.ylabel("Cumulative variance %")
plt.ylim(0, 1.01)
plt.grid(True)
plt.tight_layout()
plt.show()
```
